```python
from transformers import AutoProcessor
from transformers.audio_utils import load_audio
```

```python
from transformers import CohereAsrForConditionalGeneration
```

To log in to Hugging Face, you'll need an API token. If you don't already have one, create a token in your Hugging Face settings page (https://huggingface.co/settings/tokens).

In Colab, add the token to the secrets manager under the "🔑" in the left panel. Give it the name `HF_TOKEN`.

```python
from huggingface_hub import login
from google.colab import userdata

# Retrieve the HF_TOKEN from Colab secrets
hf_token = userdata.get('HF_TOKEN')

# Log in to Hugging Face
login(token=hf_token)
```

```python
revision = "refs/pr/6"
processor = AutoProcessor.from_pretrained("CohereLabs/cohere-transcribe-03-2026", revision=revision)
```

```python
# get the model
model = transformers.CohereAsrForConditionalGeneration.from_pretrained("CohereLabs/cohere-transcribe-03-2026", device_map="auto", revision=revision)
```

```python

audio = load_audio(
    "https://huggingface.co/datasets/hf-internal-testing/dummy-audio-samples/resolve/main/bcn_weather.mp3",
    sampling_rate=16000,
)

inputs = processor(audio, sampling_rate=16000, return_tensors="pt", language="en")
inputs.to(model.device, dtype=model.dtype)

outputs = model.generate(**inputs, max_new_tokens=256)
text = processor.decode(outputs, skip_special_tokens=True)
print(text)
```

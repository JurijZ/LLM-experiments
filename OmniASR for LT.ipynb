Lithuanian language

```python
from pydub import AudioSegment

audio = AudioSegment.from_file("Recording.m4a")
audio = audio.set_channels(1)     # convert to mono
audio = audio.set_frame_rate(16000)  # resample to 16 kHz
audio.export("Recording.wav", format="wav")
```

```python
audio_files = ["Recording.wav"]
lang = ["lth_Latn"]
transcriptions = pipeline.transcribe(audio_files, lang=lang)

transcriptions
```
